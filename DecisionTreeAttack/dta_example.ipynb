# Decision Tree Attack

Ce notebook présente l'algorithme Decision Tree Attack que nous avons recodé.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_digits
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split

import algodta
```

Nous allons tester notre algorithme avec le jeu de données MNIST

```python
digits = load_digits()
xtrain, xtest, ytrain, ytest = train_test_split(digits.data, digits.target, train_size = .8, shuffle=False)

dtc = DecisionTreeClassifier().fit(xtrain, ytrain)
```

```python
adversarial_sample = algodta.generate(xtest, dtc)
```

## Résultats

>Visualisons le premier chiffre de xtest, de adversarial_sample ainsi que leur prédiction par l'algorithme :

```python
plt.imshow(xtest[0].reshape(8,8), cmap='grey_r')
plt.show()
print(f"Le modèle prédit pour l'image ci dessus le chiffre {dtc.predict([xtest[0]])[0]}")
```

```python
plt.imshow(adversarial_sample[0].reshape(8,8), cmap='grey_r')
plt.show()

print(f"Le modèle prédit pour l'image ci dessus le chiffre {dtc.predict([adversarial_sample[0]])[0]}")
```

>Nous constatons que le modèle est trompé par la seconde image, qui à l'oeil humain ressemble à un 2.
